```python
import matplotlib.pyplot as plt
def draw_fruits(arr, ratio = 1):
    n = len(arr)    # n  .
    #   10  .   10     .
    rows = int(np.ceil(n/10))    # np.ceil            
    #  1    .   10.
    cols = n if rows < 2 else 10
    fig, axs = plt.subplots(rows, cols, figsize = (cols*ratio, rows*ratio), squeeze = False)
    
    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:    # n  .
                axs[i, j].imshow(arr[i*10 + j], cmap = 'gray_r') #  arr fruits[km.labels_ == 0] 3  .
                axs[i, j].axis('off')
    plt.show()
```

```python
inertia = []
for k in range(2,7) :
    km = KMeans(n_clusters = k, random_state = 42)
    km.fit(fruits_2d)
    inertia.append(km.inertia_)
plt.plot(range(2,7), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```
